```python
import openai
import litellm
from dotenv import load_dotenv

load_dotenv()

print("imported")

OPENAI_MODEL="gpt-4o-mini"
MESSAGES = [
    {
        "role": "user", 
         "content": "How an AI agent will proceed as an QA tester in the Requierement Analisys step"
    }
]

client = openai.OpenAI()
response_openai = client.chat.completions.create (
    model=OPENAI_MODEL,
    messages=MESSAGES
)

```

```python
print(response_openai.choices[0].message.content)
```

```python
response_litellm = litellm.completion(
    model=f"openai/{OPENAI_MODEL}", 
    messages=MESSAGES
)
```

```python
print(response_litellm.choices[0].message.content)
```

```python
response_litellm_claude = litellm.completion (
    model=f"anthropic/claude-haiku-4-5-20251001", 
    messages=MESSAGES
)
```

```python
print(response_litellm_claude.choices[0].message.content)
```
