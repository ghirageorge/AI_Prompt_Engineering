```python
import litellm
from dotenv import load_dotenv

load_dotenv()

OPENAI_MODEL="gpt-4o-mini"
MESSAGES = [
    {
        "role": "user", 
         "content": "How an AI agent will proceed as an QA tester in the Test Analisys step"
    }
]
```

```python
response_object = litellm.completion (
    model=OPENAI_MODEL, 
    messages=MESSAGES
)
```

```python
print(f"ID: {response_object.id}")
print(f"Model used: {response_object.model}")
print(f"Created timestamp: {response_object.created}")
print(f"Used data: {response_object.usage}")
```
