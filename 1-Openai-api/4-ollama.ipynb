```python
import litellm

MODEL_NAME = "ollama/llama3.2"
MAX_MODEL_DEFAULT = 500

print("LiteLLM lib is ready to use.")
```

```python
def get_completion(messages, model=MODEL_NAME, max_tokens=MAX_MODEL_DEFAULT):
    response = litellm.completion(
        messages=messages, 
        model=model, 
        max_tokens=max_tokens,
        temperature=0.2
    )

    return response.choices[0].message.content
print("Message ready")
```

```python
MESSAGES1 = [
    {
        "role": "user",
        "content": "Provide one example of a good team leader: actions, measures"
    }
]

get_completion(MESSAGES1)
```

```python
MESSAGES2 = [
    {
        "role": "user",
        "content": "I want to know how a leader speeks about its accomplisments -take as an example an QA team leader."
    }
]

get_completion(MESSAGES2)
```

```python
from IPython.display import display, Markdown

display(Markdown(get_completion(MESSAGES2)))
```
