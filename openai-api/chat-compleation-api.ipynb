```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

print("Libraries imported and env var loaded.")

MODEL_NAME = "gpt-4o-mini"

client = OpenAI()
        
response = client.chat.completions.create( 
    model=MODEL_NAME, 
    messages=[ { "role": "user", "content": "What is the most relevant skill for a QA tester in the AI filed?" } ] )

print(response.choices[0].message.content)
```
