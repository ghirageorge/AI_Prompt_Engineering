```python
# This simple instruction creates a "chain" of reasoning

import inspect
import litellm
from IPython.display import display, Markdown
from textwrap import dedent
from dotenv import load_dotenv

load_dotenv()

# --- Configuration & Context ---
MODEL_NAME = "openai/gpt-4o-mini"
MAX_TOKENS_DEFAULT = 1000 # More tokens for reasoning + code

# --- The "messy" code to be refactored ---
def get_user_initials(user_data_list):
    initials_list = []
    for user in user_data_list:
        if 'name' in user and len(user['name']) > 0:
            parts = user['name'].split(' ')
            initials = ''
            for part in parts:
                initials += part[0].upper()
            initials_list.append(initials)
    return initials_list

CODE_TO_REFACTOR = inspect.getsource(get_user_initials)

def get_completion(
    prompt,
    model=MODEL_NAME,
    max_tokens=MAX_TOKENS_DEFAULT,
    **kwargs
):
    parsed_messages = []

    if type(prompt) is str:
        parsed_messages = [
            {
                "role": "user",
                "content": prompt
            }
        ]
    else:
        parsed_messages = prompt

    response = litellm.completion(
        model=model,
        messages=parsed_messages,
        max_tokens=max_tokens,
        **kwargs
    )

    return response.choices[0].message.content


print("Setup complete. Helper functions and code context are ready.")
```

```python
# no chain-of-reasoning

prompt_no_chain_of_reasoning = [
            {
                "role": "user",
                "content": dedent(f"""
                Please refactor this Python code: 

                '''python {CODE_TO_REFACTOR}'''
                """)
            }
        ]
response_no_chain_of_reasoning = get_completion(prompt_no_chain_of_reasoning)
```

```python
display(Markdown(response_no_chain_of_reasoning))
```

```python
# With chain of reasoning



prompt_chain_of_reasoning = [
            {
                "role": "user",
                "content": dedent(f"""
                Please refactor this Python code: 
                Please think step by step and provide reasoning.
                
                '''python {CODE_TO_REFACTOR}'''
                """)
            }
        ]
response_chain_of_reasoning = get_completion(prompt_chain_of_reasoning)
```

```python
display(Markdown(response_chain_of_reasoning))
```

```python
# With chain of reasoning_2



prompt_chain_of_reasoning_2 = [
            {
                "role": "user",
                "content": dedent(f"""
                -Please refactor this Python code: 
                -Please think step by step and provide reasoning.
                -While solving the problem, cosider documentation, typing, readability, and others relavant aspects.
                
                '''python {CODE_TO_REFACTOR}'''
                """)
            }
        ]
response_chain_of_reasoning_2 = get_completion(prompt_chain_of_reasoning_2)
```

```python
display(Markdown(response_chain_of_reasoning_2))
```
